```python
df = df[df['price'] != 'Price on Request']
```

```python
df.price.value_counts().shape
```

```python
# Function to convert lac into crores.
def treat_price(X):
    if type(X) == float:
        return X
    else:
        if X[1] == 'Lac':
            return round(float(X[0])/100,2)
        else:
            return round(float(X[0]),2)
```

```python
# Use the treat_price function to convert the price column to one consistent unit that is in Crores.
df['price'] = df['price'].str.split(" ").apply(treat_price)
```

```python
# check for the unique values in the price_per_square feet column
pd.set_option('display.max_rows', None)
df.price_per_sqft.value_counts()
```

```python
# Check for the unique types of sq_ft written in the price_sq_ft column.
df['price_per_sqft'].str.extract(r'/(.*)$')[0].value_counts()
```

```python
# Extracting the only numerical value from the price_sq_ft column.
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace("₹", "").str.replace(",", "").str.strip().astype(float)
```

```python
# Working on the bedroom columns, bathroom and balcony.
df.bedRoom.value_counts()
```

```python
# Since there are very less nan in the bedroom col and
# also apparently the same rows have so much nan in the othere columns as well and 
# hence only took the observations which has no nan.
df = df[~df['bedRoom'].isnull()]
```

```python
df.head(5)
```

```python
df.bedRoom = df.bedRoom.str.split(" ").str.get(0).astype(int)
```

```python
df.bathroom = df.bathroom.str.split(" ").str.get(0).astype(int)
```

```python
# Working on the bedroom columns, bathroom and balcony.
df.balcony.value_counts()
```

```python
df.balcony = df.balcony.str.split(" ").str.get(0).str.replace("No", "0")
```

```python
df.additionalRoom = df.additionalRoom.str.lower()
```
