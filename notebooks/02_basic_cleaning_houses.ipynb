```python
import numpy as np
import pandas as pd
```

```python
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent

DATA_PATH = PROJECT_ROOT / "Data" / "Raw" / "houses.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(columns = ['link','property_id'], inplace = True)
```

```python
df.price.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Replace name of area column to price_per_sqft
df.rename(columns = {'rate':'price_per_sqft'}, inplace = True)
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.price.value_counts().shape
```

```python
def price_transform(X):
    if type(X) == float:
        return X
    else:
        if X[1] == 'Lac':
            return round(float(X[0]) / 100, 2)
        else:
            return round(float(X[0]), 2)
```

```python
# Use the treat_price function to convert the price column to one consistent unit that is in Crores.
df['price'] = df['price'].str.split(" ").apply(price_transform)
```

```python
df.head(1)
```

```python
df.price_per_sqft.shape
```

```python
# Check for the unique types of sq_ft written in the price_sq_ft column.
df['price_per_sqft'].str.extract(r'/(.*)$')[0].value_counts()
```

```python
df.price_per_sqft = df.price_per_sqft.str.split("/").str.get(0).str.replace("₹", "").str.replace(",",'').astype(float)
```

```python
df.head(1)
```

```python
df.drop(columns="area", axis = 1, inplace =True)
```

```python
df.insert(loc = 4, column="area", value = round((df['price'] * 10000000) / df["price_per_sqft"]))
```

```python
df.insert(loc = 1, column = "property_type", value = "house")
```

```python
df.head(1)
```

```python
df.society.value_counts()
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df.head(1)
```

```python
df.bedRoom.isnull().sum()
```

```python
df = df[~df.bedRoom.isnull()]
```

```python
df.bedRoom.value_counts()
```

```python
df.bedRoom = df.bedRoom.str.split(" ").str.get(0).astype(int)
```

```python
df.bathroom = df.bathroom.str.split(" ").str.get(0).astype(int).astype(int)
```

```python
df.balcony.value_counts()
```

```python
df.balcony = df.balcony.str.split(" ").str.get(0).str.replace("No", "0")
```

```python
df.head(1)
```

```python
df.additionalRoom.value_counts()
```

```python
df.additionalRoom.fillna("not_available", inplace = True)
```

```python
df.additionalRoom.isnull().sum()
```

```python
df.additionalRoom = df.additionalRoom.str.lower()
```

```python
df.noOfFloor.value_counts()
```

```python
df.noOfFloor = df.noOfFloor.str.split(" ").str.get(0)
```
